# Objetivo
Utilizar GridSpec para organizar subplots de diferentes tamaños en un análisis de ventas.

**Instrucciones:**
1. Genera un DataFrame de ventas con Mes, Ventas, y Gastos. Utiliza randint para generar números aleatorios enteros. Para ventas los valores deben estar entre 2000 y 5000. Para gastos, los valores debn estar entre 1000 y 4000.
2. Usa GridSpec para organizar los subplots de la siguiente manera:
  * Primer gráfico (barra, ocupando dos columnas)
  * Segundo gráfico (línea, una columna)
  * Tercer gráfico (dispersión, dos filas).
3. Agrega títulos y ajusta espacios.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Genera un DataFrame de ventas con Mes, Ventas, y Gastos. Utiliza randint para generar números aleatorios enteros.
# Para ventas los valores deben estar entre 2000 y 5000. Para gastos, los valores debn estar entre 1000 y 4000.
df = pd.DataFrame({
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'],
    'Ventas': np.random.randint(2000, 5000, size=12),
    'Gastos': np.random.randint(1000, 4000, size=12)
})

# Usa GridSpec para organizar los subplots de la siguiente manera:
# Primer gráfico (barra, ocupando dos columnas)
# Segundo gráfico (línea, una columna)
# Tercer gráfico (dispersión, dos filas).
fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(3, 3)

ax1 = fig.add_subplot(gs[0, :2])
ax2 = fig.add_subplot(gs[0, 2])
ax3 = fig.add_subplot(gs[1:, :])

ax1.bar(df['Mes'], df['Ventas'], color="purple")
ax2.plot(df['Mes'], df['Gastos'], marker='o', color='skyblue')
ax3.scatter(df['Ventas'], df['Gastos'])


# Agrega títulos y ajusta espacios.
ax1.set_title('Ventas por Mes (Barras)')
ax2.set_title('Gastos por Mes (Línea)')
ax3.set_title('Relación entre Ventas y Gastos (Dispersión)')
ax3.set_xlabel('Ventas')
ax3.set_ylabel('Gastos')

plt.tight_layout()
plt.show()
```
