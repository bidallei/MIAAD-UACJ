## 1. Operaciones básicas
**Objetivo:** Crear y manipular arrays.

**Instrucciones:**
1. Crea un array de 10 elementos con valores entre 1 y 10.
2. Calcula la media, mediana y desviación estándar de los valores.
3. Multiplica cada valor por 2.

```python
# 1. Crea un array de 10 elementos con valores entre 1 y 10.
import numpy as np
import random

array = np.random.randint(1, 11, 10)
print(f"Array: {array}")

# 2. Calcula la media, mediana, y desviación estándar de los valores.
media = np.mean(array)
mediana = np.median(array)
desviacion_estandar = np.std(array)

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Desviación estándar: {desviacion_estandar:2f}")

# 3. Multiplica cada valor por 2.
array_double = array * 2
print(f"Array multiplicado por 2: {array_double}")
```

## 2. Operaciones con DataFrames
**Objetivo:** Crear y manipular un DataFrame.

**Instrucciones:**
1. Crea un DataFrame con 3 columnas (A, B, C) y 5 filas con números aleatorios enteros.
2. Renombra las columnas a Col1, Col2, Col3.
3. Añade una nueva columna que sea la suma de las columnas Col1 y Col2
4. Elimina todas las filas donde el valor de Col3 sea menor que 10.

```python
# 1. Crea un DataFrame con 3 columnas (A, B, C) y 5 filas con números aleatorios enteros.
import pandas as pd

df = pd.DataFrame(np.random.randint(0, 100, size=(5, 3)), columns=['A', 'B', 'C'])
print("="*17)
print("Dataframe 5 X 3")
print("="*17)
print(df)

# 2. Renombra las columnas a Col1, Col2, Col3.
df.rename(columns={'A': 'Col1', 'B': 'Col2', 'C': 'Col3'}, inplace=True)
print("="*35)
print("Dataframe con columnas renombradas")
print("="*35)
print(df)

# 3. Añade una nueva columna que sea la suma de las columnas Col1 y Col2
df['Col4'] = df['Col1'] + df['Col2']
print("="*35)
print("Dataframe con nueva columna Col4")
print("="*35)
print(df)

# 4. Elimina todas las filas donde el valor de Col3 sea menor que 10.
df = df[df['Col3'] >= 10]
print("="*35)
print("Dataframe con filas eliminadas")
print("="*35)
print(df)
```

**Observaciones:**
- Se escogió el rango de 1 a 100 para los números aleatorios.
- Se borró la fila 5 debido a que existía un número menor que 10 en la Col3.
- Cada vez que se ejecuta el código se obtienen valores diferentes y es posible que no se eliminen filas o se eliminen otras.

## 3. Estadística descriptiva
**Objetivo:** Aplicar análisis estadístico básico.

**Instrucciones:**
1. Crea un DataFrame con 100 valores aleatorios entre 0 y 50.
2. Calcula la media, mediana y moda de los valores.
3. Encuentra el valor mínimo y el valor máximo del DataFrame.

```python
# 1. Crea un DataFrame con 100 valores aleatorios entre 0 y 50.
df_ex3 = pd.DataFrame(np.random.randint(0, 50, size=(100, 1)), columns=['A'])
print("="*38)
print("Dataframe 100 X 1 (primeros 5 valores)")
print("="*38)
print(df_ex3.head())

# 2. Calcula la media, mediana y moda de los valores.
print("="*20)
print(f"Media: {df_ex3.mean()['A']}")
print("="*20)
print(f"Mediana: {df_ex3.median()['A']}")
print("="*20)
print(f"Moda: {df_ex3.mode()['A'].iloc[0]}")

# 3. Encuentra el valor mínimo y el valor máximo del DataFrame.
print("="*20)
print(f"Valor mínimo: {df_ex3['A'].min()}")
print("="*20)
print(f"Valor máximo: {df_ex3['A'].max()}")
```

**Observaciones:**
- Cada vez que se ejecuta el código se obtiene diferentes valores.
- No se presenta el Dataframe completo por ser demasiado largo.

## 6. Gráfico de Barras
**Objetivo:** Visualizar datos categóricos.

**Instrucciones:**
1. Carga el dataset tips de Seaborn.
2. Realiza un gráfico de barras que muestre el total de las cuentas (total_bill) agrupadas por día.
3. Añade etiquetas a los ejes y un título.

```python
# 1. Carga el dataset tips de Seaborn.
import seaborn as sns
tips = sns.load_dataset('tips')
print("="*20)
print("Dataframe tips")
print("="*20)
print(tips.head())
print()

# 2. Realiza un gráfico de barras que muestre el total de las cuentas (total_bill) agrupadas por día.
# 3. Añade etiquetas a los ejes y un título.
total_por_dia = tips.groupby('day', observed=False)['total_bill'].sum()

plt.figure(figsize=(10, 6))
plt.bar(total_por_dia.index, total_por_dia.values, color='skyblue', edgecolor='black')
plt.xlabel('Día de la semana', fontsize=12)
plt.ylabel('Total de cuentas', fontsize=12)
plt.title('Total de cuentas por día', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Observaciones:**
- Para obtener el total de cuentas por día, se agrupan las cuentas de cada día y se suman. El gráfico muestra ese total.

## 7. Histograma
**Objetivo:** Visualizar la distribución de datos numéricos.

**Instrucciones:**
1. Genera 1000 números aleatorios con distribución normal (media = 0, desviación estándar = 1).
2. Crea un histograma para visualizar la distribución.
3. Ajusta el número de bins del histograma para que la visualización sea clara.

```python
# 1. Genera 1000 números aleatorios con distribución normal (media = 0, desviación estándar = 1).
data_ex7 = np.random.normal(loc=0, scale=1, size=1000)
print("Primeros 100 números:\n")
print(data_ex7[:100])

# 2. Crea un histograma para visualizar la distribución.
plt.figure(figsize=(10, 6))
plt.hist(data_ex7, bins=30, color='skyblue', edgecolor='white', alpha=0.7)

# 3. Ajusta el número de bins del histograma para que la visualización sea clara.
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Histograma de distribución normal', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Observaciones:**
- Al tener la media en 0 (loc = 0) y la desviación estándar en 1 (scale = 1), los valores aleatorios obtenidos se concentran-4 y 4, concentrándose la mayoría cerca del 0.
- Si observáramos solo los números, sería muy difícil verlo a golpe de vista, pero la gráfica nos permite verlo muy claramente de forma inmediata. De ahí que la representación gráfica sea una forma eficiente y rápida de reconocer algunos comportamientos de los datos.

## 9. Heatmap
**Objetivo:** Visualizar la correlación entre variables numéricas.

**Instrucciones:**
1. Usa el dataset iris de Seaborn.
2. Calcula la matriz de correlación de las variables.
3. Crea un heatmap para visualizar la matriz de correlación.

```python
# 1. Usa el dataset iris de Seaborn.
print("="*35)
print("Dataframe iris (5 primeros valores)")
print("="*35)
print(iris.head())
print()

# 2. Calcula la matriz de correlación de las variables.
columnas_numericas = iris.select_dtypes(include=['float64', 'int64'])
correlation_matrix = columnas_numericas.corr()

# 3. Crea un heatmap para visualizar la matriz de correlación.
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de correlación de variables numéricas', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 10. Gráfico de Línea con Media Móvil
**Objetivo:** Visualizar series temporales.

**Instrucciones:**
1. Genera una serie temporal de 100 días con valores aleatorios.
2. Calcula la media móvil de los últimos 10 días y grafícala junto a la serie original.
3. Usa diferentes colores para la serie original y la media móvil.

```python
# 1. Genera una serie temporal de 100 días con valores aleatorios.
fechas = pd.date_range(start='2025-08-01', periods=100, freq='D')
serie_temporal = pd.Series(np.random.rand(100), index=fechas)
print("="*35)
print("Serie temporal (primeras 10 filas)")
print("="*35)
print(serie_temporal.head(10))
print()

# 2. Calcula la media móvil de los últimos 10 días y grafícala junto a la serie original.
media_movil = serie_temporal.rolling(window=10).mean()
print("="*35)
print("Media móvil (primeros 10 valores)")
print("="*35)
print(media_movil[10:20])
print()

# 3. Usa diferentes colores para la serie original y la media móvil.
plt.figure(figsize=(10, 6))
plt.plot(serie_temporal, label='Serie temporal', color='blue')
plt.plot(media_movil, label='Media móvil', color='red')
plt.xlabel('Tiempo', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.title('Serie temporal y media móvil', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='both', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Observaciones:**
- La media móvil se obtiene a partir del décimo día.
- En la gráfica se observa que la media móvil tiene un comportamiento más o menos similar a los números aleatorios.

## 11. Datos Temporales con Análisis Estadístico
**Objetivo:** Trabajar con series temporales y calcular estadísticas.

**Instrucciones:**
1. Genera una serie temporal de 200 días con valores aleatorios.
2. Calcula la media y la desviación estándar de la serie temporal.
3. Visualiza la serie junto con las bandas de desviación estándar (es decir, la
media ±1 desviación estándar).

```python
# 1. Genera una serie temporal de 200 días con valores aleatorios.
fechas = pd.date_range(start='2024-04-01', periods=200, freq='D')
serie_temporal = pd.Series(np.random.rand(200), index=fechas)

# 2. Calcula la media y la desviación estándar de la serie temporal.
media = serie_temporal.mean()
desviacion_estandar = serie_temporal.std()
print(f"Media: {media:.4f}")
print(f"Desviación estándar: {desviacion_estandar:.4f}")

# 3. Visualiza la serie junto con las bandas de desviación estándar (es decir, la media ±1 desviación estándar).
plt.figure(figsize=(10, 6))
plt.plot(serie_temporal, label='Serie temporal', color='blue')
plt.fill_between(serie_temporal.index, media - desviacion_estandar, media + desviacion_estandar, color='gray', alpha=0.3, label='Desviación estándar')
plt.xlabel('Tiempo', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.axhline(y=media, color='red', linestyle='--', alpha=0.8, linewidth=1, label=f'Media: {media:.3f}')
plt.title('Serie temporal con bandas de desviación estándar', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='both', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Observaciones:**
- La gráfica muestra que la serie se mueve alrededor de la media y que casi todos los valores están dentro del rango de la media, así que tiene un comportamiento más o menos estable.
