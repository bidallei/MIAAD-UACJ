```python
#Carga y exploración del dataset:

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Cargar el dataset Iris utilizando seaborn o cualquier otro dataset de su elección.
df = sns.load_dataset('iris')

#Verificar las primeras filas del dataset con el método .head().
print("===Dataset: Iris===")
print(df.head())
print()

#Obtener información general sobre los datos (nombres de columnas, tipos de datos, valores nulos, etc.).
print("===Información general del dataset Iris:===")
print(df.info())
```

```python
# Crear los siguientes gráficos con Matplotlib:
# a) Gráfico de barras:
# - Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.
# - Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

petal_stats = df.groupby('species')[['petal_length', 'petal_width']].mean()
petal_stats.plot(kind='bar', color= ['green', 'red'])
plt.title('Comparación de promedios de largo y ancho de pétalos', fontsize="14")
plt.xlabel('Especie')
plt.ylabel('Centímetros')
plt.show()
```

```python
# Crear los siguientes gráficos con Matplotlib:
# b) Histograma:
# Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).
# Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.
plt.figure(figsize=(8, 6))
plt.hist(df['sepal_length'], bins=20, color='darkolivegreen', edgecolor='dimgrey')
plt.title('Distribución de los largos de los sépalos')
plt.xlabel('Largo en centímetros')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Crear los siguientes gráficos con Matplotlib:
# c) Gráfico de dispersión (scatter plot):
# Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
# Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.
plt.scatter(df['petal_length'], df['petal_width'], c=df['species'].map({'setosa': 'yellow', 'versicolor': 'gray', 'virginica': 'red'}), alpha=0.5)
plt.title('Relación entre el largo del pétalo y el ancho del pétalo')
plt.xlabel('Largo del pétalo en centímetros')
plt.ylabel('Ancho del pétalo en centímetros')
plt.show()
```

```python
# Crear los siguientes gráficos con Matplotlib:
# d) Gráfico de cajas (box plot):
# Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).
# Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.
labels = ['setosa', 'versicolor', 'virginica']

setosa_sl = df[df['species'] == 'setosa']['sepal_length']
versicolor_sl = df[df['species'] == 'versicolor']['sepal_length']
virginica_sl = df[df['species'] == 'virginica']['sepal_length']

plt.boxplot([setosa_sl, versicolor_sl, virginica_sl], tick_labels=labels)
plt.title('Distribución del largo del sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Largo del sépalo en centímetros')
plt.show()

setosa_sw = df[df['species'] == 'setosa']['sepal_width']
versicolor_sw = df[df['species'] == 'versicolor']['sepal_width']
virginica_sw = df[df['species'] == 'virginica']['sepal_width']

plt.boxplot([setosa_sw, versicolor_sw, virginica_sw], tick_labels=labels)
plt.title('Distribución del ancho del sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Ancho del sépalo en centímetros')
plt.show()
```

```python
# Crear los siguientes gráficos con Matplotlib:
# e) Gráfico de líneas:
# Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.
# Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['petal_length'], color='red')
plt.title('Tendencia de la longitud del pétalo a lo largo del dataset')
plt.xlabel('Índice de observación')
plt.ylabel('Largo del pétalo en centímetros')
plt.grid(True)
plt.show()
```
