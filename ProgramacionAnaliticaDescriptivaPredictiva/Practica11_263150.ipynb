# Objetivo
Utilizar Seaborn junto con plt.subplots() para visualizar diferentes aspectos de un conjunto de datos en una sola figura.

**Instrucciones:**
* Carga el dataset "penguins" de Seaborn. Utiliza df = sns.load_dataset("penguins")
* Usa plt.subplots() para crear una figura con 2 filas y 2 columnas.
* Genera los siguientes gráficos:
  * (0,0): sns.histplot(): Histograma de las longitudes de aletas.
  * (0,1): sns.boxplot(): Boxplot comparando la masa corporal por especie.
  * (1,0): sns.scatterplot(): Dispersión entre masa corporal y largo de aletas.
  * (1,1): sns.violinplot(): Violin plot de la masa corporal según la especie.
* Personaliza los gráficos con títulos y colores adecuados.
* Ajusta el diseño con plt.tight_layout().

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Carga el dataset "penguins" de Seaborn. Utiliza df = sns.load_dataset("penguins")
df = sns.load_dataset("penguins")

#Usa plt.subplots() para crear una figura con 2 filas y 2 columnas.
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#Genera los siguientes gráficos:
# (0,0): sns.histplot(): Histograma de las longitudes de aletas.
sns.histplot(data=df, x='flipper_length_mm', ax=axs[0, 0], color='#FF4653')

# (0,1): sns.boxplot(): Boxplot comparando la masa corporal por especie.
sns.boxplot(data=df, x='species', y='body_mass_g', ax=axs[0, 1], color='#00A896')

# (1,0): sns.scatterplot(): Dispersión entre masa corporal y largo de aletas.
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', ax=axs[1, 0], color='#7D8CA3')

# (1,1): sns.violinplot(): Violin plot de la masa corporal según la especie.
sns.violinplot(data=df, x='species', y='body_mass_g', ax=axs[1, 1], color='#721817')

# Personaliza los gráficos con títulos y colores adecuados.
fig.suptitle('Análisis del dataset <Pingüinos>', fontsize=16, fontweight='bold')
axs[0, 0].set_title('Histograma de las longitudes de aletas')
axs[0, 1].set_title('Boxplot comparando la masa corporal por especie')
axs[1, 0].set_title('Dispersión entre masa corporal y largo de aletas')
axs[1, 1].set_title('Violin plot de la masa corporal según la especie')

# Ajusta el diseño con plt.tight_layout().
plt.tight_layout()
plt.show()
```
